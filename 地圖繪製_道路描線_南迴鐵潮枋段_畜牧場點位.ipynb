```python
from folium import plugins
import folium
import os
import pandas as pd
import numpy as np
#from pykml import parser
```

```python
m = folium.Map([22.492115, 120.511789], zoom_start=15)

## 選定座標位置，及起始縮放大小
```

```python
m
```

```python
df = pd.read_excel('屏東縣畜牧場位置.xlsx', sheet_name=1)
```

```python
df.head()
```

```python
#df.info()
```

```python
畜牧場座標 = []
for i, j in zip(df['緯度'], df['經度']):
    畜牧場座標.append((i, j))
```

```python
#畜牧場座標
```

```python
for coord in 畜牧場座標:
    test_m = folium.Circle(location=[coord[0], coord[1]], color='red', radius=200, fill=True, fill_opacity=0.5).add_to(m)
```

```python
folium.Marker(
    location=[22.369528, 120.592583],
    popup='人和橋',
    icon=folium.Icon(color='blue', icon='fa-tint', prefix='fa')
).add_to(m)

## 南寮大道及榮濱路路口

folium.Marker(
    location=[22.388191, 120.580323],
    popup='南旗尾橋',
    icon=folium.Icon(color='blue', icon='fa-tint', prefix='fa')
).add_to(m)

## 台15線及延平路路口

folium.Marker(
    location=[22.465532, 120.511481],
    popup='牛埔溪橋',
    icon=folium.Icon(color='blue', icon='fa-tint', prefix='fa')
).add_to(m)

folium.Marker(
    location=[22.470051, 120.515769],
    popup='牛埔溪橋上游',
    icon=folium.Icon(color='blue', icon='fa-tint', prefix='fa')
).add_to(m)

folium.Marker(
    location=[22.494722, 120.511667],
    popup='溪洲溪橋',
    icon=folium.Icon(color='blue', icon='fa-tint', prefix='fa')
).add_to(m)

folium.Marker(
    location=[22.495192, 120.509741],
    popup='溪洲溪橋上游',
    icon=folium.Icon(color='blue', icon='fa-tint', prefix='fa')
).add_to(m)
```
