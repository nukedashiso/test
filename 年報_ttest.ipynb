```python
import numpy as np
import pandas as pd
```

```python
import scipy.stats
```

```python
df = pd.read_excel('final地下水_Sylvia 1091229.xlsx', sheet_name='表3(作圖)')
```

```python
df.head()
```

```python
def t_test_new(place):
    
    group11 = list(pd.to_numeric(df[place].iloc[0:3], errors='coerce').fillna(value=0))
    group22 = list(pd.to_numeric(df[place].iloc[3:28], errors='coerce').fillna(value=0))
    
    mean1 = np.mean(group11)
    mean2 = np.mean(group22)
    std1 = np.std(group11)
    std2 = np.std(group22)
    nobs1 = len(group11)
    nobs2 = len(group22)
    
    modified_std1 = np.sqrt(np.float32(nobs1) / np.float32(nobs1-1)) * std1
    modified_std2 = np.sqrt(np.float32(nobs2) / np.float32(nobs2-1)) * std2
    statistic, pvalue = scipy.stats.ttest_ind_from_stats( 
               mean1=mean1, std1=modified_std1, nobs1=nobs1,   
               mean2=mean2, std2=modified_std2, nobs2=nobs2 )
    return statistic, pvalue

# 參考資料：https://medium.com/@peilee_98185/t-%E6%AA%A2%E5%AE%9A-with-python-443c2364b071
```

```python
name_of_place = ['隧道出水口', '土方堆置A區', '土方堆置B區',             #水位
                 '隧道出水口.1', '土方堆置A區.1', '土方堆置B區.1',       #水溫
                 '隧道出水口.2', '土方堆置A區.2', '土方堆置B區.2',       #pH
                 '隧道出水口.3', '土方堆置A區.3', '土方堆置B區.3',       #BOD
                 '隧道出水口.4', '土方堆置A區.4', '土方堆置B區.4',       #COD
                 '隧道出水口.5', '土方堆置A區.5', '土方堆置B區.5',       #TOC
                 '隧道出水口.6', '土方堆置A區.6', '土方堆置B區.6',       #TDS
                 '隧道出水口.7', '土方堆置A區.7', '土方堆置B區.7',       #SS
                 '隧道出水口.8', '土方堆置A區.8', '土方堆置B區.8',       #NH4-N
                 '隧道出水口.9', '土方堆置A區.9', '土方堆置B區.9',       #導電度
                 '隧道出水口.10', '土方堆置A區.10', '土方堆置B區.10',    #比導電度
                 '隧道出水口.11', '土方堆置A區.11', '土方堆置B區.11',    #SO42-
                 '隧道出水口.12', '土方堆置A區.12', '土方堆置B區.12',    #NO3-
                 '隧道出水口.13', '土方堆置A區.13', '土方堆置B區.13',    #總菌落數
                 '隧道出水口.14', '土方堆置A區.14', '土方堆置B區.14',    #大腸桿菌群
                 '隧道出水口.15', '土方堆置A區.15', '土方堆置B區.15',    #Fe2+
                 '隧道出水口.16', '土方堆置A區.16', '土方堆置B區.16',    #Mn2+
                 '隧道出水口.17', '土方堆置A區.17', '土方堆置B區.17',    #油脂
                 '隧道出水口.18', '土方堆置A區.18', '土方堆置B區.18',]   #Cl-
```

```python
list__11 = []
for place in name_of_place:
    list__11.append(t_test_new(place))
```

```python
df_tp = pd.DataFrame(list__11, index=name_of_place , columns=['T-Score', 'P-Value'])
```
