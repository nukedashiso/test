```python
def 歷季逐時風速(filename, sheetname):
    #list_of_windspeed = []
    df = pd.read_excel(filename, sheet_name=sheetname, header=None)
    df.columns = df.iloc[0, :]
    df.drop(index=[0, 1, 26, 27, 28, 29, 30, 31, 32, 33, 34], inplace=True)
    #list_of_windspeed.append(list(df['風速']))
    return list(df['風速'])

def 歷季逐時最頻風向(filename, sheetname):
    df = pd.read_excel(filename, sheet_name=sheetname, header=None)
    df.columns = df.iloc[0, :]
    df.drop(index=[0, 1, 26, 27, 28, 29, 30, 31, 32, 33, 34], inplace=True)
    #list_of_windspeed.append(list(df['風速']))
    return list(df['最頻風向'])
```

```python
list_of_windspeed_曾文5號橋 = []
for filename in filenames:
    list_of_windspeed_曾文5號橋 += 歷季逐時風速(filename, '曾文5號橋')
    print(filename, list_of_windspeed_曾文5號橋)
    
list_of_winddirection_曾文5號橋 = []
for filename in filenames:
    list_of_winddirection_曾文5號橋 += 歷季逐時最頻風向(filename, '曾文5號橋')
    print(filename, list_of_winddirection_曾文5號橋)
```

```python
list_of_windspeed_青年活動中心 = []
for filename in filenames:
    list_of_windspeed_青年活動中心 += 歷季逐時風速(filename, '青年活動中心')
    print(filename, list_of_windspeed_青年活動中心)
    
list_of_winddirection_青年活動中心 = []
for filename in filenames:
    list_of_winddirection_青年活動中心 += 歷季逐時最頻風向(filename, '青年活動中心')
    print(filename, list_of_winddirection_青年活動中心)
```

```python
list_of_windspeed_楠西區照興里 = []
for filename in filenames:
    list_of_windspeed_楠西區照興里 += 歷季逐時風速(filename, '楠西區照興里')
    print(filename, list_of_windspeed_楠西區照興里)
    
list_of_winddirection_楠西區照興里 = []
for filename in filenames:
    list_of_winddirection_楠西區照興里 += 歷季逐時最頻風向(filename, '楠西區照興里')
    print(filename, list_of_winddirection_楠西區照興里)
```
