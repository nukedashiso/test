```python
mpl.rcParams.update({'font.size': 12})
plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率
## 改變圖中所有字體大小
```

```python
from windrose import WindroseAxes, WindAxes, plot_windrose
```

```python
#df1.head()
```

```python
def draw_windrose(df, output_name):
    ax = WindroseAxes.from_ax()
    ax.bar(df['風向'], df['風速'], bins=np.arange(0.0, 5.0, 0.5), normed=True, cmap=cm.coolwarm)
    
    ax.set_yticks(np.arange(0, 25, 5))
    ax.set_yticklabels(np.arange(0, 25, 5))
    ax.yaxis.set_major_formatter(ticker.PercentFormatter())
    
    Leg = plt.legend(loc='right', bbox_to_anchor=(1.35, 0.2), title='wind speed(m/s)')
    i = 0.0
    j = 0
    while i<5.0:
        Leg.get_texts()[j].set_text(str(i) + '~' + str(i+0.5))
        i += 0.5
        j += 1
        
    plt.grid(c='k')
    
    plt.savefig(output_name, dpi=1200)
    plt.show()
    
## 風化圖繪製def，下面有單步驟繪製說明

## 輸出圖片檔Legend被切掉，再研究是否可把savefig圖片全存
```

```python
draw_windrose(df1, '曾文5號橋_def_test.png')
```
